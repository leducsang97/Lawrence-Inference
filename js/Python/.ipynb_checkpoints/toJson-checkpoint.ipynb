```python
import pandas as pd
import json
```

```python
df = pd.read_csv("faculty.csv", encoding="latin")
```

```python
df
```

```python
jsonFile = []
```

```python
for d in df.iterrows():
    j = {"name":d[1]["First Name"]+" "+d[1]["Last Name"],
        "description":d[1]["Interests"],
        "additional":{"photo":d[1]["Photo Link"],"department":d[1]["Department"]},
        }
    jsonFile.append(j)
```

```python
y = json.dumps(jsonFile)
```
